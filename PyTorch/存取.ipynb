```python
import torch
import matplotlib.pyplot as plt
from torch.autograd import Variable
#(100,1)数据unsqueeze一维变二维
#dim dim=0时（1,100），dim=1时（100,1）
x=torch.unsqueeze(torch.linspace(-1,1,100),dim=1)
#rand返回的是0到1的均匀分布,下面返回（100,1）数据
y=x.pow(2)+0.2*torch.rand(x.size())

x,y=Variable(x,requires_grad=False),Variable(y,requires_grad=False)

def save():
    net1=torch.nn.Sequential(torch.nn.Linear(1,10),torch.nn.ReLU(),torch.nn.Linear(10,1))
    optimizer=torch.optim.SGD(net1.parameters(),lr=0.2)
    loss_func=torch.nn.MSELoss()
    print(111)
    #训练
    for i in range(500):
        prediction=net1(x)
        loss=loss_func(prediction,y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    #绘制
    plt.figure(1, figsize=(10, 3))
    p1=plt.subplot(1,3,1)
    p1.scatter(x.data.numpy(),y.data.numpy())
    p1.set_title('net1')
    p1.plot(x.data.numpy(),prediction.data.numpy(),'r',lw=2)
    #保存整个网络
    torch.save(net1,'net.pkl')

    #保存网络中的参数(速度快, 占内存少)
    torch.save(net1.state_dict(),'net_params.pkl')

    
```

```python
#提取网络
def restore_net():
    net2=torch.load('net.pkl')
    prediction=net2(x)
    
    p2=plt.subplot(1,3,2)
    p2.set_title('restore_net')
    p2.scatter(x.data.numpy(),y.data.numpy())
    p2.plot(x.data.numpy(),prediction.data.numpy(),'r',lw=1)

    
def restore_params():
    net3=torch.nn.Sequential(torch.nn.Linear(1,10),torch.nn.ReLU(),torch.nn.Linear(10,1))
    #将参数赋值到net3
    net3.load_state_dict(torch.load('net_params.pkl'))
    prediction=net3(x)
    
    p3= plt.subplot(1,3,3)
    p3.set_title('params')
    p3.scatter(x.data.numpy(),y.data.numpy())
    p3.plot(x.data.numpy(),prediction.data.numpy(),'r',lw=2)
    plt.show()
    
save()
# plt.figure(1,figsize = (8, 4))
# restore entire net (may slow)
restore_net()

# restore only the net parameters
restore_params()
```
