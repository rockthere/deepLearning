```python
#导入
import matplotlib.pyplot as plt
import numpy as np
#-1到1之间产生50个数据
x=np.linspace(-1, 1, 50)
y=x**2+1

#plt.figure定义图像窗口
plt.figure()
plt.plot(x,y)
#显示
plt.show()
```

```python
x=np.linspace(-1, 1, 50)
y1=x**2+1
y2=2*x+1

#定义一个图像窗口，编号3，大小（8,5）
plt.figure(num=3,figsize=(8,5))
plt.plot(x,y1)

#设置x轴坐标范围
plt.xlim((-2,2))

#设置y轴坐标范围
plt.ylim(-2,4)

#设置x轴坐标名称
plt.xlabel('X')
#设置y轴坐标名称
plt.ylabel('Y')


#设置x轴刻度及名称
new_xticks = np.linspace(-2, 2, 7)
new_yticks = np.linspace(-2, 4, 10)
plt.xticks(new_xticks)
#设置y轴刻度,以及刻度名称
plt.yticks([-2, 0, 1, 2,3,4],[r'start -2',0,1,2,3,4])

plt.plot(x,y2,color='r',linewidth=1.0,linestyle='--')
plt.show()
```

```python
plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-2, 2))
plt.ylim((-2, 4))

#获取当前坐标轴信息
ax=plt.gca()
#设置右边框，隐藏
ax.spines['right'].set_color('none')
#设置上边框隐藏
ax.spines['top'].set_color('none')

#设置x轴坐标轴刻度的位置,位置包含（top，bottom，both，default，none）
#设置y轴set_ticks_position（left，right，both，default，none）
ax.xaxis.set_ticks_position('bottom')

#设置边框位置(outward，axes，data)
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))


plt.show()
```

```python
#标注
x=np.linspace(-3,3,50)
y=1.5*x+1

plt.figure(num=1,figsize=(8,5))
plt.plot(x,y)


#移动门坐标
ax=plt.gca()
#隐藏右边边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#设置x轴刻度位置
ax.xaxis.set_ticks_position('bottom')

#x轴线移动到y=0位置
ax.spines['bottom'].set_position(('data',0))

ax.spines['left'].set_position(('data',0))


x0=1
y0=1.5*x0+1
#点（x0,0）到点(x0,y0)的线，颜色红色虚线
plt.plot([x0,x0],[0,y0],'r--',linewidth=2)

#绘制一个点
plt.scatter([x0],[y0],s=100,color='b')

#添加标注annotate,xycoords=‘data’基于数据的值选择位置。xytext设置偏移量，textcoords设置偏移方式，fontsize设置字体大小，arrowprops设置箭头类型
plt.annotate('1.5x+1=%s'%y0,xy=(x0,y0),xycoords='data',xytext=(+100,-30), textcoords='offset points', fontsize=16,arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))


#添加text注释(-2,4)字体绘制的起始点坐标，fontdict字体设置
plt.text(-2,4,'This is text',fontdict={'size':16,'color':'r'})
plt.show()
```

```python
#散点图绘制

#绘制点数
n=300
#均值0，方差1，的标准正态分布
x=np.random.normal(0,1,n)
y=np.random.normal(0,1,n)
T=np.arctan2(y,x)

#xy，对应坐标，s点的大小，c点的颜色，alpha透明度
plt.scatter(x,y,s=70,c=T,alpha=0.8)
plt.show()

```

```python
#柱状图
n=12
x=np.arange(n)
# print(x)
y1=(1-x/float(n))*np.random.uniform(0.5,1.0,n)
y2=(1-x/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(x,y1)
plt.bar(x,-y2)

#x,y坐标范围
plt.xlim(-0.5,n)
plt.ylim(-1.25,1.25)
#隐藏x,y坐标刻度值
plt.xticks(())
plt.yticks(())

#设置颜色,edgecolor边框颜色，facecolor  bar颜色
plt.bar(x,y1,facecolor='r',edgecolor='w')
plt.bar(x,-y2,facecolor='g',edgecolor='w')


#绘制y值
for x0,y0 in zip(x,y1):
    plt.text(x0+0.1,y0+0.2,'%.2f' % y0,ha='center',va='bottom')

for x0,y0 in zip(x,y2):
    plt.text(x0+0.1,-y0-0.2,'%.2f' % y0,ha='center',va='bottom')

plt.show()
```

```python
#等高校线图

#定义高
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n=256
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
#meshgrid生成网格采样点的函数，返回的两个矩阵行和列是相同的，且X,Y的行数都等于输入参数y中元素的总个数。X,Y列数都等于x元素的总个数
X,Y=np.meshgrid(x,y)

#绘制等高线
plt.contour(X,Y,f(X,Y),8,colors='black')

#添加颜色,8表示等高线的密集程度，0表示一分为2
#cmap是颜色图谱，具体值可以参考http://www.cnblogs.com/denny402/p/5122594.html
c=plt.contourf(X,Y,f(X,Y),8,alpha=.89,cmap=plt.cm.hot)

plt.clabel(c, inline=True, fontsize=15)

#清楚坐标
plt.xticks(())
plt.yticks(())

plt.show()
```

```python

#随机矩阵画图
a = np.array([0.313660827978, 0.365348418405, 0.423733120134,
              0.365348418405, 0.439599930621, 0.525083754405,
              0.423733120134, 0.525083754405, 0.651536351379]).reshape(3,3)
plt.imshow(a, interpolation='nearest', cmap='bone', origin='lower')
#右边显示的不同颜色的值，shrink可以指定高度为原高度的百分比
plt.colorbar(shrink=.8)
plt.show()
```

```python
#3D图像绘制
#导入
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
#在窗口上添加3D坐标
ax = Axes3D(fig)

#初始化数据
x=np.arange(-4,4,0.25)
y=np.arange(-4,4,0.25)

#网格化
X,Y=np.meshgrid(x, y)#x,y平面网格

#定义Z轴
Z=np.sin(np.sqrt(X**2+Y**2))

#rstride 和 cstride 分别代表 row 和 column 的跨度。值越大，网格间距越大
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
#zdir表示投影，值可以x,y,z
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))

plt.show()
```

```python
#多图绘制

plt.figure()
x=np.arange(-1,4,0.05)
y1=2*x+0.5
y2=x**2
y3=x**3-x**2
y4=x**2-x

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.text(1.5,np.max(y1)+1,s='2x+0.5')

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)


plt.show()
```

```python

#不均匀绘制
#第一行绘制一个
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.text(1.5,np.max(y1)+1,s='2x+0.5')

#重新调整，2行，3列标号是4（此时把第一行的当成3个对待）
plt.subplot(2,3,4)
plt.plot(x,y2)

plt.subplot(2,3,5)
plt.plot(x,y3)

plt.subplot(2,3,6)
plt.plot(x,y4)


plt.show()
```

```python
#分格显示
plt.figure()
#(3, 3)3行3列，（0,0）图的绘制起始位置，colspan列的跨度，此处设置3，表示在（3,3）中占据3列，
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.plot(x,y1)    # 画小图
ax1.set_title('2x+0.5')  # 设置小图的标题


ax2=plt.subplot2grid((3,3),(1,0),rowspan=2)
ax2.plot(x,y1)


ax3=plt.subplot2grid((3,3),(1,1),colspan=2)
ax3.plot(x,y2)

ax4 = plt.subplot2grid((3, 3), (2, 1))
ax5 = plt.subplot2grid((3, 3), (2, 2))

#画两个点
ax5.scatter([1, 2], [2, 2])


plt.show()
```

```python
#gridspec
import matplotlib.gridspec as gridspec

plt.figure()
#3行3列（从0开始）
gs = gridspec.GridSpec(3, 3)
#0,:表示占据0行和所有列
ax1=plt.subplot(gs[0,:])
ax1.plot(x,y1)

#1,:2表示绘制，第一行，及第2列前的所有行
ax2=plt.subplot(gs[1,:2])
ax2.plot(x,y1)

plt.show()
```

```python
#subplots

#sharex是否共享x轴坐标，sharey是否共享y轴坐标
f, ((ax11, ax12), (ax13, ax14)) = plt.subplots(2, 2, sharex=False, sharey=True)

ax11.plot(x,y1)
plt.show()
```
