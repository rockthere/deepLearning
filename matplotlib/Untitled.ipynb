```python
#某个城市11点到12点每分钟的温度折线图  15-18度
#数据准备
x=range(60)
y_hangzhou=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]
#创建画布
plt.figure(figsize=(20,8),dpi=80)
#绘制图像
#linestyle -实线 --虚线 -.点划线  :点虚线 
plt.plot(x,y_hangzhou,color='red',linestyle='-.',label="杭州")
plt.plot(x,y_beijing,label="北京")

#显示图例
plt.legend(loc='upper right')

# 设置x,y轴刻度range(40)[::5]或range(0，40，5)
#刻度说明
x_label=["11点{}分".format(i) for i in x]
#添加网格 linestyle='--'虚线,alpha透明度
plt.grid(linestyle='--',alpha=0.5)

#添加说明
plt.xlabel("时间变化")
plt.ylabel("温度变化")
plt.title("杭州北京11点到12点温度变化折线图")
plt.yticks(range(40)[::5])
plt.xticks(x[::5],x_label[::5])
# 显示图像
plt.show()
```

```python
#某个城市11点到12点每分钟的温度折线图  15-18度
#数据准备
x=range(60)
y_hangzhou=[random.uniform(15,18) for i in x]
y_beijing=[random.uniform(1,3) for i in x]
#创建画布
#plt.figure(figsize=(20,8),dpi=80)
figure,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)
#绘制图像
#linestyle -实线 --虚线 -.点划线  :点虚线 
axes[0].plot(x,y_hangzhou,color='red',linestyle='-.',label="杭州")
axes[1].plot(x,y_beijing,label="北京")

#显示图例
axes[0].legend(loc='upper right')
axes[1].legend(loc='upper right')

# 设置x,y轴刻度range(40)[::5]或range(0，40，5)
#刻度说明
x_label=["11点{}分".format(i) for i in x]


#添加网格 linestyle='--'虚线,alpha透明度
axes[0].grid(linestyle='--',alpha=0.5)
axes[1].grid(linestyle='--',alpha=0.5)

#添加说明

axes[0].set_xticks(x[::10])
axes[0].set_xticklabels(x_label[::10])
axes[0].set_yticks(range(40)[::5])


axes[1].set_xticks(x[::10])
axes[1].set_xticklabels(x_label[::10])
axes[1].set_yticks(range(40)[::5])

axes[0].set_xlabel("时间变化")
axes[0].set_ylabel("温度变化")
axes[0].set_title("杭州11点到12点温度变化折线图")

axes[1].set_xlabel("时间变化")
axes[1].set_ylabel("温度变化")
axes[1].set_title("北京11点到12点温度变化折线图")

# 显示图像
plt.show()
```

```python
#柱状图
x=["第一天","第二天","第三天"]
y=[40,50,25]

plt.figure()
x_ticks=range(len(x))
plt.bar(x_ticks,y,color=['r','g','b'])
plt.xticks(x_ticks,x)
plt.title("每天收入")
plt.grid()
plt.show
```

```python
#柱状图
x=["张三","李四","王二"]
y=[40,50,25]
y1=[100,80,90]

plt.figure()
x_ticks=range(len(x))
plt.bar(x_ticks,y,width=0.2,label='首日')
plt.bar([i+0.2 for i in x_ticks],y1,width=0.2,label="首周")
plt.legend()
plt.xticks(x_ticks,x)
plt.title("每天收入")
plt.grid()
plt.show
```

```python
#直方图
#组距 （最大-最小）/组数
time=[2,4,3,3,13,13,6,7,8 ,10]
plt.figure()
distance=2
#y轴表示的频率或者频数
group_num=int(max(time)-min(time)/distance)
plt.hist(time,bins=group_num,density=True)
plt.show()
```

```python
#饼图 pie()x:数量，自动计算百分比 labels:每部分名称 sutopct占比显示指定 %1.2f%% colors:每份颜色
x=["第一次","第二次","第三次","第四次"]
num=[10,100,20,30]
plt.figure()
plt.pie(num,labels=x,colors=['r','g','b','y'])

plt.axis("equal")
plt.show()
```
