```python
#pandas图表 数据处理工具 panel 数据面板  data

import pandas as pd
import numpy as np
# DataFrame可以行列索引，二维数组处理
data=np.random.normal(0,1,(10,5))

data=pd.DataFrame(data)
#添加索引行
stock=["股票{}".format(i) for i in range(10)]
pd.DataFrame(data,index=stock)

#添加列索引
date=pd.date_range(start="20200101",periods=5,freq="B")

pd.DataFrame(data,index=stock,columns=date)

print(data.shape)
print(data.index)
print(data.columns)
print(data.values)

print(data.head(3))
print(data.tail(2))

#修改索引
stock=["股票-{}".format(i) for i in range(10)]
data.index=stock


#重设索引
data.reset_index()#多一列
data.reset_index(drop=True)
```

```python
df=pd.DataFrame({'month':[1,4,7,10],
                 'year':[2018,2019,2020,2021],
                 'sale':[55,40,84,31]})

#set_index(keys,drop=True) 以某列值设置为新的索引 keys列索引或者列索引名称列表 
#drop是否当做新的索引删除原来列
print(df.set_index("month",drop=False))


#设置多个索引，以年和月份
new_df=df.set_index(["year","month"])

#MultiIndex
print(new_df.index)
#Panel是DataFrame容器
```
