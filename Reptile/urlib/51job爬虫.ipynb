```python
#51job爬虫学习
import re
import requests
key=input("请输入")
# https://search.51job.com/list/080200,000000,0000,00,9,99,java,2,1.html
#?lang=c&postchannel=0000&workyear=99&cotype=99&degreefrom=99&jobterm=99&companysize=99&ord_field=0&dibiaoid=0&line=&welfare=
data={"lang":"c",
      "postchannel":"0000",
      "workyear":"99",
      "cotype":"99","degreefrom":"99","jobterm":"99","companysize":"99","ord_field":"0","dibiaoid":"0"}
hd={"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36"}
response=requests.get("https://search.51job.com/list/080200,000000,0000,00,9,99,"+key+",2,1.html",headers=hd,params=data)
print(response.encoding)
response.encoding="gbk"
pat_allnum="共(.*?)条职位"
pat_pages='<span class="td">共(.*?)页'

#也可以使用总数量除以每页数量得到总页数
allnum=re.compile(pat_allnum,re.S).findall(response.text)[0]
pages=(int)(re.compile(pat_pages,re.S).findall(response.text)[0])

print(pages)

for i in range(0,pages):
    print("----------正在爬"+str(i+1)+"页----------")
    reItem=requests.get("https://search.51job.com/list/080200,000000,0000,00,9,99,"+key+",2,"+str(i+1)+".html",headers=hd,params=data)
    reItem.encoding="gbk"

    pat_url='\)"></em>.*?<a target="_blank" .*?href="(.*?)"  onmousedown='
    urls=re.compile(pat_url,re.S).findall(reItem.text)
    
    for job_url in urls:
        #print(job_url)

        jobRes=requests.get(job_url)
        jobRes.encoding="gbk"
        pat_title='<h1 title="(.*?)">'
        pat_company='<p class="cname">.*?title="(.*?)"'
        pat_msg='<div class="bmsg job_msg inbox">(.*?)<div class="share">'
        pat_money='<div class="tHeader tHjob">.*?<strong>(.*?)</strong>'
        title=re.compile(pat_title,re.S).findall(jobRes.text)
        company=re.compile(pat_company,re.S).findall(jobRes.text)
        msg=re.compile(pat_msg,re.S).findall(jobRes.text)
        money=re.compile(pat_money,re.S).findall(jobRes.text)

        print(money)
        print(title)
        print(company)
        print("------------------------------")



```
